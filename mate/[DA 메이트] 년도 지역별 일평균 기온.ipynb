문자열 -> Datetime으로 변환
- strptime으로 변환
- Datetime에서 년,월,일,시,분 추출

```python
temp_12["일시"] = temp_12["일시"].apply(lambda x : datetime.strptime(x,'%Y-%m-%d %H:%M'))
```

```python
temp_12["년"] = temp_12["일시"].apply(lambda x : x.year)
temp_12["월"] = temp_12["일시"].apply(lambda x : x.month)
temp_12["일"] = temp_12["일시"].apply(lambda x : x.day)
temp_12["시"] = temp_12["일시"].apply(lambda x : x.hour)
temp_12["분"] = temp_12["일시"].apply(lambda x : x.minute)
```

### 3의 배수인 시간대만 추출하기

환자발생정보가 어느시간에 나오는 지 알려주지 않고, 하루에 몇 명 발생했는지 알려주기 때문에, 시간대별 기온을 사용하는 것은 의미없어 보여 일평균기온을 사용

일평균기온에는 여러가지 방법이 있지만 현재 우리나라는 3시간 간격으로 관측한 8회 관측값(3, 6, 9, 12, 15, 18, 21, 24시)의 평균 기온 사용

- 행마다 하나씩 for문에 넣어주는 방식
- 3의 배수인 시간대의 index만 추출하기
- temp_12_3h : 3의 배수인 시간대만 추출한 것

```python
idx_lst = []

for idx,row in temp_12.iterrows():
    if row["시"] % 3 == 0:
        idx_lst.append(idx)
```

```python
temp_12_3h = temp_12.loc[idx_lst,:]
temp_12_3h
```

### 시도별로 추출하기
- a = 강원도
- 강릉, 대관령, 동해, 북강릉, 속초, 영월, 원주, 인제, 정선군, 철원, 춘천, 태백,홍천
- 시도별로 해당되는 지점의 데이터 추출
- 해당 지점들의 기온을 통해 평균기온 구하기

```python
def 시도별일평균기온(시도,data):

    # 딕셔너리 형태로 시도별 지점 정보저장
    loc_dic = {"강원도":["강릉","대관령","동해","북강릉","속초","영월","원주","인제","정선군","철원","춘천","태백","홍천"],
              "경기도":["동두천","수원","양평","이천","파주"]}
    
    # 최종데이터
    data_lst = []
    
    for loc in loc_dic[시도]:
        data_lst.append(data[data["지점명"] == loc])
        
    df = pd.concat(data_lst,axis=0,ignore_index = True)
    
    # 지점별 일별로 기온평균 내기
    
    mean_df = df.groupby(["지점","지점명","년","월","일"]).mean().reset_index()
    
    return mean_df.drop(["시","분"],axis=1)
```
